```python
import pandas as pd

# 1. Load the data
df = pd.read_csv('/content/GlobalWeatherRepository.csv')

# 2. Automatic Aggregation (Finding Averages)
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
country_summary = df.groupby('country')[numeric_cols].mean()

# 3. Show Results
print("Country-wise Average Weather Data:")
print(country_summary.head(10))
```
